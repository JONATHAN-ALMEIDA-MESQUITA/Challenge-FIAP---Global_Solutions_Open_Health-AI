## **Introduo sobre sade e diabete**

O Sistema de Vigilncia de Fatores de Risco Comportamentais (BRFSS)  uma pesquisa telefnica relacionada  sade coletada anualmente pelo CDC. A cada ano, a pesquisa obtm respostas de mais de 400.000 americanos sobre comportamentos de risco relacionados  sade, condies crnicas de sade e uso de servios preventivos. A pesquisa  realizada anualmente desde 1984. Para este projeto, foi utilizado um arquivo CSV do conjunto de dados disponvel no Kaggle para o ano de 2015. Este conjunto de dados original contm respostas de 441.455 indivduos e possui 330 caractersticas. Essas caractersticas so ou perguntas feitas diretamente aos participantes ou variveis calculadas com base nas respostas individuais dos participantes.

 um conjunto de dados limpo com 253.680 respostas a uma pesquisa da BRFSS2015 do CDC. A varivel-alvo Diabetes_012 tem 3 classes. 0 significa sem diabetes ou apenas durante a gravidez, 1  para pr-diabetes e 2  para diabetes. H um desequilbrio de classes neste conjunto de dados. Ele possui 21 variveis de caractersticas.

**Colunas:**

Nome em Ingls, (Nome em Portugus)
  
  Descrio em Portugus

1. **Diabetes_012** (Diabetes_012) - [0, 1, 2] - ['Classe 0', 'Classe 1', 'Classe 2']
  "Diabetes_012"  uma varivel em um conjunto de dados sobre diabetes. Com trs classes distintas, ela categoriza os estados relacionados  diabetes da seguinte forma:

    **Classe 0:** Representa a ausncia de diabetes ou diabetes apenas durante a gravidez.

    **Classe 1:** Indica pr-diabetes.

    **Classe 2:** Representa diabetes.

  Essa varivel  til para analisar e entender diferentes nveis e tipos de diabetes em um conjunto de dados especfico. Cada classe oferece insights sobre o estado de sade relacionado  diabetes para os indivduos includos no conjunto de dados.

2. **Presso Sangunea Alta** (HighBP) - [0, 1] - ['No', 'Sim']

  A presso sangunea alta, ou hipertenso,  uma condio em que a fora do sangue contra as paredes das artrias  consistentemente muito alta. A hipertenso  um fator de risco para vrias condies de sade, incluindo doenas cardiovasculares.

3. **Nveis elevados de colesterol** (HighChol) - [0, 1] - ['No', 'Sim']

  Se refere a nveis elevados de colesterol no sangue. O colesterol  uma substncia gordurosa que o corpo precisa para construir clulas, mas nveis elevados podem aumentar o risco de doenas cardacas e outros problemas de sade.

4. **Exame de colesteral** (CholCheck) - [0, 1] - ['No', 'Sim']

  Se refere a um teste ou exame de colesterol, como um check-up dos nveis de colesterol no sangue, ento h uma relao indireta com a diabetes. Pessoas com diabetes, especialmente diabetes tipo 2, tm um risco aumentado de desenvolver doenas cardacas. O controle adequado dos nveis de colesterol  uma parte importante do manejo global para reduzir o risco cardiovascular em pessoas com diabetes.

5. **ndice de Massa Corporal** (BMI) - [...]

  O ndice de Massa Corporal (IMC)  uma medida que relaciona o peso e a altura de uma pessoa. Ele  frequentemente utilizado como uma ferramenta para avaliar se uma pessoa tem um peso saudvel, est abaixo do peso, com sobrepeso ou obesidade. O IMC  calculado dividindo o peso da pessoa em quilogramas pelo quadrado da altura em metros (IMC = peso em kg / altura em metros^2).
  
  A relao entre o IMC e o diabetes est relacionada ao fato de que o excesso de peso, especialmente a obesidade,  um fator de risco significativo para o desenvolvimento de diabetes tipo 2. Pessoas com IMC mais elevado tm um maior risco de resistncia  insulina, uma condio na qual as clulas do corpo no respondem adequadamente  insulina, levando a nveis elevados de glicose no sangue.

6. **Fumante** (Smoker) - [0, 1] - ['No', 'Sim']

  Fumar  especialmente prejudicial para pessoas com diabetes devido aos riscos aumentados de complicaes cardiovasculares, circulatrias, neurolgicas e renais. O tabagismo pode comprometer ainda mais o controle glicmico e agravar problemas de sade associados ao diabetes.

7. **Acidente Vascular Cerebral** (Stroke) - [0, 1] - ['No', 'Sim']

  Pessoas com diabetes tm um risco aumentado de sofrer um acidente vascular cerebral (AVC) devido  associao entre diabetes e doenas cardiovasculares. O diabetes pode causar danos aos vasos sanguneos, aumentando o risco de complicaes cerebrovasculares. O controle adequado da diabetes, juntamente com a gesto de fatores de risco como presso arterial elevada e colesterol,  crucial para reduzir a probabilidade de AVC em indivduos com diabetes. Consultar regularmente um profissional de sade  fundamental para um manejo eficaz.

8. **Doena Cardaca** (HeartDiseaseorAttack) - [0, 1] - ['No', 'Sim']

  Pessoas com diabetes tm mais chances de ter problemas no corao, como doenas cardacas e ataques cardacos. A diabetes prejudica os vasos sanguneos, eleva a presso arterial e aumenta o risco de acumulao de placas nas artrias. Controlar a diabetes e fatores de risco, como presso e colesterol,  crucial para diminuir essas chances. Consultar um profissional de sade regularmente ajuda a gerenciar esses riscos.

9. **Atividade Fsica** (PhysActivity) - [0, 1] - ['No', 'Sim']

Alm disso, o exerccio regular contribui para o bem-estar geral, reduzindo o estresse e melhorando a qualidade do sono. Antes de iniciar um programa de exerccios,  recomendvel consultar um profissional de sade para garantir que seja seguro e adequado s necessidades individuais.

Controle as pores para evitar picos nos nveis de glicose. Evite sucos de frutas processados, pois podem ter alto teor de acar. ### Analise de habitos e sade por grupos: Sem diabetes, Diabetes, Pr-diabetes



```python
# Mapeamento de novos nomes para as variveis
novos_nomes = {
    'HighBP': 'Presso Sangunea Alta',
    'HighChol': 'Nveis Elevados de Colesterol',
    'PhysActivity': 'Atividade Fsica',
    'Fruits': 'Consumo de Frutas',
    'Veggies': 'Consumo de Vegetais',
    'HvyAlcoholConsump': 'Consumo Pesado de lcool'
}

# Renomeie as variveis no DataFrame
dfs[0] = dfs[0].rename(columns=novos_nomes)

# Mapeamento entre valores numricos e rtulos
mapeamentos = {
    'Presso Sangunea Alta': {0: 'No', 1: 'Sim'},
    'Nveis Elevados de Colesterol': {0: 'No', 1: 'Sim'},
    'Atividade Fsica': {0: 'No', 1: 'Sim'},
    'Consumo de Frutas': {0: 'No', 1: 'Sim'},
    'Consumo de Vegetais': {0: 'No', 1: 'Sim'},
    'Consumo Pesado de lcool': {0: 'No', 1: 'Sim'}
}

# Aplica os mapeamentos para cada varivel
for variavel, mapeamento in mapeamentos.items():
    dfs[0][variavel] = dfs[0][variavel].map(mapeamento)


```

```python
# Lista de variveis de interesse atualizada
variaveis_interesse = ['Presso Sangunea Alta', 'Nveis Elevados de Colesterol', 'Atividade Fsica',
                       'Consumo de Frutas', 'Consumo de Vegetais', 'Consumo Pesado de lcool']

# Calcule o nmero de linhas e colunas com base no nmero de variveis
num_variaveis = len(variaveis_interesse)
num_colunas = 2  # Coloque quantos grficos voc deseja em cada linha

# Calcule o nmero de linhas necessrio (dividindo por 2, arredondando para cima)
num_linhas = -(-num_variaveis // num_colunas)  # Tcnica para arredondamento para cima em Python

# Crie subplots para cada par de variveis
fig, axs = plt.subplots(num_linhas, num_colunas, figsize=(15, 10))

# Itere sobre os pares de variveis
for i, variavel in enumerate(variaveis_interesse):
    # Calcule a posio do subplot
    linha = i // num_colunas
    coluna = i % num_colunas

# Crie um grfico de barras empilhadas para cada par de variveis
    sns.countplot(x=variavel, hue='Diabetes_012', data=dfs[0], ax=axs[linha, coluna])

    # Adicione as porcentagens nas barras
    total = len(dfs[0][variavel])
    for p in axs[linha, coluna].patches:
        height = p.get_height()
        axs[linha, coluna].text(p.get_x() + p.get_width() / 2., height + 0.01, f'{height / total:.1%}', ha="center")

    axs[linha, coluna].set_title(variavel)
    axs[linha, coluna].set_xlabel('')
    axs[linha, coluna].set_ylabel('Contagem')

# Ajuste o layout
plt.tight_layout()
plt.show()
```
