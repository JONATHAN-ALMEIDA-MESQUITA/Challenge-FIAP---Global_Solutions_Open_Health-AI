## **Introdução sobre saúde e diabete**

O Sistema de Vigilância de Fatores de Risco Comportamentais (BRFSS) é uma pesquisa telefônica relacionada à saúde coletada anualmente pelo CDC. A cada ano, a pesquisa obtém respostas de mais de 400.000 americanos sobre comportamentos de risco relacionados à saúde, condições crônicas de saúde e uso de serviços preventivos. A pesquisa é realizada anualmente desde 1984. Para este projeto, foi utilizado um arquivo CSV do conjunto de dados disponível no Kaggle para o ano de 2015. Este conjunto de dados original contém respostas de 441.455 indivíduos e possui 330 características. Essas características são ou perguntas feitas diretamente aos participantes ou variáveis calculadas com base nas respostas individuais dos participantes.


### **DF[0] diabetes_012_health**

É um conjunto de dados limpo com 253.680 respostas a uma pesquisa da BRFSS2015 do CDC. A variável-alvo Diabetes_012 tem 3 classes. 0 significa sem diabetes ou apenas durante a gravidez, 1 é para pré-diabetes e 2 é para diabetes. Há um desequilíbrio de classes neste conjunto de dados. Ele possui 21 variáveis de características.


---
**Colunas:**

Nome em Inglês, (Nome em Português)
  
  Descrição em Português

1. **Diabetes_012** (Diabetes_012) - [0, 1, 2] - ['Classe 0', 'Classe 1', 'Classe 2']
  "Diabetes_012" é uma variável em um conjunto de dados sobre diabetes. Com três classes distintas, ela categoriza os estados relacionados à diabetes da seguinte forma:

    **Classe 0:** Representa a ausência de diabetes ou diabetes apenas durante a gravidez.

    **Classe 1:** Indica pré-diabetes.

    **Classe 2:** Representa diabetes.

  Essa variável é útil para analisar e entender diferentes níveis e tipos de diabetes em um conjunto de dados específico. Cada classe oferece insights sobre o estado de saúde relacionado à diabetes para os indivíduos incluídos no conjunto de dados.

2. **Pressão Sanguínea Alta** (HighBP) - [0, 1] - ['Não', 'Sim']

  A pressão sanguínea alta, ou hipertensão, é uma condição em que a força do sangue contra as paredes das artérias é consistentemente muito alta. A hipertensão é um fator de risco para várias condições de saúde, incluindo doenças cardiovasculares.

3. **Níveis elevados de colesterol** (HighChol) - [0, 1] - ['Não', 'Sim']

  Se refere a níveis elevados de colesterol no sangue. O colesterol é uma substância gordurosa que o corpo precisa para construir células, mas níveis elevados podem aumentar o risco de doenças cardíacas e outros problemas de saúde.

4. **Exame de colesteral** (CholCheck) - [0, 1] - ['Não', 'Sim']

  Se refere a um teste ou exame de colesterol, como um check-up dos níveis de colesterol no sangue, então há uma relação indireta com a diabetes. Pessoas com diabetes, especialmente diabetes tipo 2, têm um risco aumentado de desenvolver doenças cardíacas. O controle adequado dos níveis de colesterol é uma parte importante do manejo global para reduzir o risco cardiovascular em pessoas com diabetes.

5. **Índice de Massa Corporal** (BMI) - [...]

  O Índice de Massa Corporal (IMC) é uma medida que relaciona o peso e a altura de uma pessoa. Ele é frequentemente utilizado como uma ferramenta para avaliar se uma pessoa tem um peso saudável, está abaixo do peso, com sobrepeso ou obesidade. O IMC é calculado dividindo o peso da pessoa em quilogramas pelo quadrado da altura em metros (IMC = peso em kg / altura em metros^2).
  
  A relação entre o IMC e o diabetes está relacionada ao fato de que o excesso de peso, especialmente a obesidade, é um fator de risco significativo para o desenvolvimento de diabetes tipo 2. Pessoas com IMC mais elevado têm um maior risco de resistência à insulina, uma condição na qual as células do corpo não respondem adequadamente à insulina, levando a níveis elevados de glicose no sangue.

6. **Fumante** (Smoker) - [0, 1] - ['Não', 'Sim']

  Fumar é especialmente prejudicial para pessoas com diabetes devido aos riscos aumentados de complicações cardiovasculares, circulatórias, neurológicas e renais. O tabagismo pode comprometer ainda mais o controle glicêmico e agravar problemas de saúde associados ao diabetes.

7. **Acidente Vascular Cerebral** (Stroke) - [0, 1] - ['Não', 'Sim']

  Pessoas com diabetes têm um risco aumentado de sofrer um acidente vascular cerebral (AVC) devido à associação entre diabetes e doenças cardiovasculares. O diabetes pode causar danos aos vasos sanguíneos, aumentando o risco de complicações cerebrovasculares. O controle adequado da diabetes, juntamente com a gestão de fatores de risco como pressão arterial elevada e colesterol, é crucial para reduzir a probabilidade de AVC em indivíduos com diabetes. Consultar regularmente um profissional de saúde é fundamental para um manejo eficaz.

8. **Doença Cardíaca** (HeartDiseaseorAttack) - [0, 1] - ['Não', 'Sim']

  Pessoas com diabetes têm mais chances de ter problemas no coração, como doenças cardíacas e ataques cardíacos. A diabetes prejudica os vasos sanguíneos, eleva a pressão arterial e aumenta o risco de acumulação de placas nas artérias. Controlar a diabetes e fatores de risco, como pressão e colesterol, é crucial para diminuir essas chances. Consultar um profissional de saúde regularmente ajuda a gerenciar esses riscos.

9. **Atividade Física** (PhysActivity) - [0, 1] - ['Não', 'Sim']

  A atividade física é essencial para pessoas com diabetes. Ela ajuda a controlar os níveis de glicose, gerenciar o peso, melhorar a saúde cardiovascular e aumentar a sensibilidade à insulina. Além disso, o exercício regular contribui para o bem-estar geral, reduzindo o estresse e melhorando a qualidade do sono. Antes de iniciar um programa de exercícios, é recomendável consultar um profissional de saúde para garantir que seja seguro e adequado às necessidades individuais.

10. **Consumo Frutas** (Fruits) - [0, 1] - ['Não', 'Sim']

  Pessoas com diabetes podem comer frutas, mas é importante fazer escolhas conscientes. Opte por frutas frescas e inteiras, com casca, e prefira aquelas com baixo índice glicêmico e alto teor de fibras. Controle as porções para evitar picos nos níveis de glicose. Evite sucos de frutas processados, pois podem ter alto teor de açúcar. Consulte um profissional de saúde para orientações personalizadas sobre a inclusão de frutas em sua dieta.

11. **Consumo Vegetais** (Veggies) - [0, 1] - ['Não', 'Sim']

  Vegetais são fundamentais para pessoas com diabetes. Eles são ricos em nutrientes, fibras e geralmente têm baixo índice glicêmico. Consumir uma variedade de vegetais coloridos é recomendado. Opte por opções não processadas, controle as porções e prefira métodos de cozimento saudáveis. Consultar um profissional de saúde ajuda a personalizar a inclusão de vegetais na dieta de acordo com as necessidades individuais.

12. **Consumo Pesado de Álcool** (HvyAlcoholConsump) - [0, 1] - ['Não', 'Sim']

  O consumo pesado de álcool está associado a um aumento no risco de desenvolver diabetes tipo 2. O álcool pode afetar a regulação da glicose no sangue, contribuindo para a resistência à insulina. Além disso, o consumo excessivo de álcool pode levar a problemas de saúde, como obesidade e doenças hepáticas, que são fatores de risco adicionais para diabetes. Portanto, é aconselhável limitar o consumo de álcool e manter um estilo de vida saudável para prevenir complicações relacionadas à diabetes. Consultar um profissional de saúde é crucial para orientação personalizada.

13. **Assistência Médica** (AnyHealthcare) - [0, 1] - ['Não', 'Sim']
  Assistência médica para diabetes envolve cuidados regulares para gerenciar a condição. O envelhecimento também pode afetar o manejo da diabetes, exigindo ajustes nas estratégias de tratamento. É importante que pessoas de todas as idades estejam cientes dos fatores de risco e mantenham um estilo de vida saudável para prevenir ou gerenciar a diabetes. ### Analise de habitos e saúde por grupos: Sem diabetes, Diabetes, Pré-diabetes



```python
# Mapeamento de novos nomes para as variáveis
novos_nomes = {
    'HighBP': 'Pressão Sanguínea Alta',
    'HighChol': 'Níveis Elevados de Colesterol',
    'PhysActivity': 'Atividade Física',
    'Fruits': 'Consumo de Frutas',
    'Veggies': 'Consumo de Vegetais',
    'HvyAlcoholConsump': 'Consumo Pesado de Álcool'
}

# Renomeie as variáveis no DataFrame
dfs[0] = dfs[0].rename(columns=novos_nomes)

# Mapeamento entre valores numéricos e rótulos
mapeamentos = {
    'Pressão Sanguínea Alta': {0: 'Não', 1: 'Sim'},
    'Níveis Elevados de Colesterol': {0: 'Não', 1: 'Sim'},
    'Atividade Física': {0: 'Não', 1: 'Sim'},
    'Consumo de Frutas': {0: 'Não', 1: 'Sim'},
    'Consumo de Vegetais': {0: 'Não', 1: 'Sim'},
    'Consumo Pesado de Álcool': {0: 'Não', 1: 'Sim'}
}

# Aplica os mapeamentos para cada variável
for variavel, mapeamento in mapeamentos.items():
    dfs[0][variavel] = dfs[0][variavel].map(mapeamento)


```

```python
# Lista de variáveis de interesse atualizada
variaveis_interesse = ['Pressão Sanguínea Alta', 'Níveis Elevados de Colesterol', 'Atividade Física',
                       'Consumo de Frutas', 'Consumo de Vegetais', 'Consumo Pesado de Álcool']

# Calcule o número de linhas e colunas com base no número de variáveis
num_variaveis = len(variaveis_interesse)
num_colunas = 2  # Coloque quantos gráficos você deseja em cada linha

# Calcule o número de linhas necessário (dividindo por 2, arredondando para cima)
num_linhas = -(-num_variaveis // num_colunas)  # Técnica para arredondamento para cima em Python

# Crie subplots para cada par de variáveis
fig, axs = plt.subplots(num_linhas, num_colunas, figsize=(15, 10))

# Itere sobre os pares de variáveis
for i, variavel in enumerate(variaveis_interesse):
    # Calcule a posição do subplot
    linha = i // num_colunas
    coluna = i % num_colunas

# Crie um gráfico de barras empilhadas para cada par de variáveis
    sns.countplot(x=variavel, hue='Diabetes_012', data=dfs[0], ax=axs[linha, coluna])

    # Adicione as porcentagens nas barras
    total = len(dfs[0][variavel])
    for p in axs[linha, coluna].patches:
        height = p.get_height()
        axs[linha, coluna].text(p.get_x() + p.get_width() / 2., height + 0.01, f'{height / total:.1%}', ha="center")

    axs[linha, coluna].set_title(variavel)
    axs[linha, coluna].set_xlabel('')
    axs[linha, coluna].set_ylabel('Contagem')

# Ajuste o layout
plt.tight_layout()
plt.show()
```
